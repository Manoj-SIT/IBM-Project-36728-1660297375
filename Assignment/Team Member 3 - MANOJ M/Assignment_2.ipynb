```python
#Check for Categorical columns
df_categorical = df[['Gender', 'Age', 'IsActiveMember', 'Exited', 'EstimatedSalary']]
df_categorical.head()
```

```python
#Perfrom ENCODING
from sklearn.preprocessing import LabelEncoder
salary = LabelEncoder()
salary.fit(df_categorical['EstimatedSalary'])
s = salary.transform(df_categorical['EstimatedSalary'])
print("Before Encoding:", list(df_categorical['EstimatedSalary'][-10:]))
print("After Encoding:", s[-10:])
print("The inverse from the encoding result:", salary.inverse_transform(s[-10:]))
```
